```python
import geocoder

def get_pincode(latitude, longitude):
    # Combine latitude and longitude
    location = geocoder.osm([latitude, longitude], method='reverse')
    
    # Extract pincode from the location
    pincode = location.postal
    
    return pincode

# Example usage
latitude = 13.0396882
longitude = 77.5568631
pincode = get_pincode(latitude, longitude)
print("Pincode:", pincode)
```
